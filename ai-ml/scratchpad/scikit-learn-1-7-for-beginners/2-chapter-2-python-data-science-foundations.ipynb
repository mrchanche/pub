### Working with numpy arrays

NumPy = Numerical Python

numpy arrays are extremely fast and that makes it useful for handling large data.numpy also has built in data handling tools for calculations, reshaping etc...


```python
import numpy as np

# 1D array (vector)
a = np.array([1,2,3,4,5])

print("1D array:", a)

# 2D array (matrix)
b = np.array([[1,2,3],[4,5,6]])
print("2D array:\n", b)

# Array attributes and shape
print("Shape of b:", b.shape) # 2 rows, 3 columns
print("Data type:", a.dtype) # int64 float32 etc...

```

```python
# Common Operations

c = np.array([10, 20, 30, 40, 50])
print("Element-wise addition", a + c)
print("Element-wise multiplication:", a * 2)

# Mathametical functions
print("Mean:", a.mean())
print("Sum:", a.sum())
print("Max:", a.max())
print("Square root:", np.sqrt(a))
```

```python
# Indexing and Slicing

print("First element:", a[0])
print("Elements 2 to 4:", a[1:4])
print("Second row of b:", b[1])
print("First column of b:", b[:,0])
```

```python
# Reshaping Arrays (converting 1d to 2d or 2d to 1d etc...)

d = np.arange(12) # creates 0 - 11
d2 = d.reshape(3, 4) # 3 rows, 4 columns
print("Original array:", d)
print("Reshaped array:\n", d2)
```

```python
# Random Arrays

random_arr = np.random.rand(5) # 5 random floats (0 to 1)
random_matrix = np.random.randn(3, 3) # 3x3 matrix normal distribution
print("Random array:", random_arr)
print("random matrix:\n", random_matrix)
```

### 2.3 Data Manipulation with Pandas

Swiss army knife of data science.

Load, clean, transform and explore data. 

Pandas allows you to easily get data in to a DataFrame to shape your data. It can filter, aggregate and transform 1000's of lines of messy data with just a few lines of code.

```python
import pandas as pd

# DataFrames are like spreadsheets, rows and columns, each with a label

data = {
    "Name":["Alice", "Bob", "Cathy", "Dan"],
    "Age":[25, 30, 22, 35],
    "Score":[88, 92, 95, 85]
}

df = pd.DataFrame(data) # Pass the data to the pd DataFrame
print(df)

# Save to CSV
df.to_csv('sample_data.csv', index=False)
```

### Reading Data From Files

```python
df = pd.read_csv("sample_data.csv")
print(df.head()) # Show the first 5 rows
print("\n**************\n")
print(df.info()) # quick summary
```

```python
# Inspecting and Exploring Data

print("Columns:", df.columns)
print("\n")
print("Shape:", df.shape) # (rows, columns)
print("\n")
print("Summary stats:\n", df.describe())
print("\n")
print("First 3 rows:\n", df.head(3))
print("\n")
print("Unique values in 'Age':", df["Age"].unique())
```

### Selecting and filtering data

```python
# Select a column
ages = df["Age"]

# Select multiple columns
subset = df[["Name","Score"]]

# Page 42
# Select rows by index
row2 = df.iloc[1] # Second row (0-based index)

# Filter rows by condition
adults = df[df["Age"] >= 25]
print(adults)
```

### Adding, Modifying, and Removing Data

```python
# Add a new column
df["Passed"] = df["Score"] >= 90
print(df,"\n")
# Modify values
df.loc[df["Name"] == "Dan", "Score"] = 90
print(df,"\n")
# Remove a column
df = df.drop("Passed", axis=1)
print(df,"\n")
# Remove rows with missing values
df = df.dropna()
print(df,"\n")
# Fill missing values
df["Age"] = df["Age"].fillna(df["Age"].mean())
print(df,"\n")
```

### Sorting and Grouping

```python
# Sort by column
sorted_df = df.sort_values("Score", ascending = False)
print(sorted_df)
```

```python
# Group and summarize
avg_scores = df.groupby("Age")["Score"].mean()
print(avg_scores)

# Grouping and aggregating data is very useful for discovering trends and preparing features for machine learning.
```

### Practical Example - Cleaning a Dataset

```python
# If you have a messy CSV and need data cleaned up

df = pd.read_csv("students.csv")
print(df.isnull().sum())    # Check for missing values
df["Score"] = df["Score"].fillna(df["Score"].medium())
df = df[df["Age"] >= 18] # Keep only adults
df = df.drop_duplicates() # Remove duplicate rows
print(df.head())
```

### Exporting Data

```python
df.to_csv("clearned_data.csv", index=False)
```

### 2.4 Data Visualization with Matplotlib

Page 46

Clear visuals make your insights on data accessible to others. Matplotlib works with numpy and pandas.

Plotting data will often reveal errors and patterns. 

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100) # 100 points between 0 and 10
y = np.sin(x)

plt.plot(x,y)
plt.title("Sine Wave")
plt.xlabel("x value")
plt.ylabel("sin(x)")
plt.show()
```

### Bar and Histogram Plots

```python
# Page 47

categories = ["A", "B", "C", "D"]
values = [23, 45, 56, 78]

plt.bar(categories, values)
plt.title("Category Counts")
plt.ylabel("Count")
plt.show()
```

```python
# Histograms

data = np.random.randn(1000) # 1000 random numbers (normal distribution)
plt.hist(data, bins = 30)
plt.title("Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequence")
plt.show()
```

### Scatter Plots for Relationships

```python
x = np.random.randn(50)
y = 2 * x + np.random.randn(50) * 0.1 # Line with some noise

plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

### Visualizing DataFrames

### Customizing Your Plots

```python
plt.plot(x, y, color="purple", linestyle="--", marker="o", label="Sine Curve")
plt.title("Customized Line Plot")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
plt.grid(True)
plt.show()
```
