```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Example data, sq m house and corresponding price
X = np.array([[50],[60],[80],[100],[120]]) # Features, house sizes
y = np.array([150, 180, 240, 300, 360]) # Labels, prices

# Create and train the model
model = LinearRegression()
model.fit(X,y)

# Predict price for a new home of 90 sq m
new_house_size = np.array([[90]])

predicted_price = model.predict(new_house_size)
print(f"Predicted price for a 90 sq m house: ${predicted_price[0]:.0f}k")
```

The alogrithm learned that 90 sq m should equate to 270k.

Nearly every task in scikit follows this process

1. Import the relevant class
2. Create an instance of the model
3. Fit the model to training data `.fit()`
4. Predict outcomes on new data `.predict()`
5. Evaluate model performance

```python
# Page 17 Classify Iris Flowers

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Load dataset
iris = load_iris()
X = iris.data # Features (measurements)
y = iris.target # Lavels (species)

# Split data into train/test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Initialize the model
knn = KNeighborsClassifier(n_neighbors=3)

# Fit (train) the model
knn.fit(X_train, y_train)

# Predict on new data
predictions = knn.predict(X_test)

# Evaluate accuracy
accuracy = knn.score(X_test, y_test)
print(f"Model accuracy: {accuracy:.2f}")
```

The advantage here is going from test/train data to a model with predictions. Scikit learn gives you the pre-built scaffolding.

```python
# Import libs
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# datasets, access sample data
# train_test_split self explanatory
# LogisticRegression this is the ML model

# Load the dataset
digits = datasets.load_digits()
print("Data Shape:", digits.data.shape) # Features
print("Target Shape:", digits.target.shape) # Labels
```

```python
# Look at the digits
import matplotlib.pyplot as plt

plt.gray()
plt.matshow(digits.images[0])
plt.show()
print("Label:", digits.target[0])
```

```python
# Split the data into training and test sets

X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=42) # 20% test size, 42 = seed number

print("Training Samples:", X_train.shape[0])
print("Test samples:", X_test.shape[0])
```

```python
# Create a logistic regression classifier and train

model = LogisticRegression(max_iter = 2000)
model.fit(X_train, y_train)
```

```python
# Make predictions

y_pred = model.predict(X_test)
print("First 10 predictions:", y_pred[:10])
print("First 10 true labels:", y_test[:10])
```

```python
# Evaluate performance

score = model.score(X_test, y_test)
print(f"Accuracy: {score:.2f}")
```

We Loaded a real dataset, split it into train and test sets, then trained an ML model, made predictions and checked for accuracy.
